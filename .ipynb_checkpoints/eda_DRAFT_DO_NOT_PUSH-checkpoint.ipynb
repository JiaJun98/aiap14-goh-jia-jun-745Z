### Retreiving data from database

```python
#Establishing connection with sqlite3
conn = sqlite3.connect('data/failure.db')
```

```python
# Execute a SELECT statement
failure_df = pd.read_sql_query("SELECT * FROM failure", conn)

#A copy is made as it is indepedent of the dataframe, modifications made to copy will not affect main vice versa
failure_df_copy = failure_df.copy()
```

### Exploratory Data Analysis

### Data Cleaning and Transformation

### Checking for missing values

```python
failure_df_copy.isnull().sum()
#From the results, there are 384 values that are null value
```

```python
#Measuring percentage of missing values from total dataset
print(f"Total dataset: {len(failure_df_copy)}")
print(f"Percentage of total dataset: {384/len(failure_df_copy)*100:.2f}%")
```

```python
#Since the missing dataset is only 3.8 % of the whole dataset, it is safe to remove the missing values
failure_df_copy_no_missing = failure_df_copy.dropna(axis=0, inplace=False)
```

### Understanding of variables

```python
failure_df_copy_no_missing.shape
#This shows that there is 9697 rows and 14 columns
```

```python
#This shows all the unique values of each column.
#As you can see each type of Failure (Depedendent variable) is a binary label as there is only 2 labels
failure_df_copy_no_missing.nunique(axis=0)
```

```python
#Car ID is likely to just an ID and redudant. The column can be removed from the dataset
failure_df_copy_no_missing_no_id = failure_df_copy_no_missing.drop('Car ID', axis=1, inplace=False)
```

### Numerical Variables

```python
failure_df_copy_no_missing_no_id.describe()
#As you can see from the results, the descriptive statistics show the central tendency, dispersion, and shape.
#Initially, there seems to be an error for RPM as the min < 0. We should investigate the percentage of negative values
#for RPM
```

```python
failure_df_copy_no_missing_no_id.query("RPM < 0")
```

```python
#1.From above, there is 214 rows that has negative RPM. On first sight, it seems errorneous but negative RPM
#suggests that there is negative torque speed. As torque speed is a vector, it sussgesst the car engine is rotating
#at the other direction

#2.We can also observe that the temperature is both in degrees celsius and farenheit. 
#There is a need to keep the units in temperature consistent

#3. We can also observe for the column Model, the value is Model <ModelNumber>, Year. It should be split into 2 columns

#4. We can also observe that for the column Factory, city and country is also combined. It should also be split into 2 columns
```

```python
#Thus for 2) the following transfomration is done
failure_df_copy_no_missing_no_id[['Temperature_Value', 'Temperature_Unit']] = failure_df_copy_no_missing_no_id['Temperature'].str.split(' ', expand=True)
failure_df_copy_no_missing_no_id.loc[failure_df_copy_no_missing_no_id['Temperature_Unit'] == '°F', 'Temperature_Value'] = (failure_df_copy_no_missing_no_id.loc[failure_df_copy_no_missing_no_id['Temperature_Unit'] == '°F', 'Temperature_Value'].astype(float) - 32) * 5/9
failure_df_copy_no_missing_no_id.loc[failure_df_copy_no_missing_no_id['Temperature_Unit'] == '°F', 'Temperature_Unit'] = '°C'
failure_df_copy_no_missing_no_id['Temperature'] = failure_df_copy_no_missing_no_id['Temperature_Value'].astype(float).round(1).astype(str)
failure_df_copy_no_missing_no_id_constant_temp = failure_df_copy_no_missing_no_id.drop(['Temperature_Value', 'Temperature_Unit'], axis=1)
failure_df_copy_no_missing_no_id_constant_temp["Temperature"] = failure_df_copy_no_missing_no_id_constant_temp["Temperature"].apply(lambda x: float(x))
```

```python
#After convertion: Note since all values are in °C, there is no need for the unit anymore
failure_df_copy_no_missing_no_id_constant_temp.query("RPM < 0")
```

```python
#Splitting Model into Model Name and Year Column- need to carry out some moving adjustments
modelID = failure_df_copy_no_missing_no_id_constant_temp.columns.get_loc('Model')
failure_df_copy_no_missing_no_id_constant_temp[['Model', 'Year']] = failure_df_copy_no_missing_no_id_constant_temp['Model'].str.split(', ', expand=True)
failure_df_copy_no_missing_no_id_constant_temp.insert(modelID + 1, 'Year', failure_df_copy_no_missing_no_id_constant_temp.pop('Year'))
```

```python
#Splitting Factory into City and Country 
FactoryID = failure_df_copy_no_missing_no_id_constant_temp.columns.get_loc('Factory')
failure_df_copy_no_missing_no_id_constant_temp[['City', 'Country']] = failure_df_copy_no_missing_no_id_constant_temp['Factory'].str.split(', ', expand=True)
failure_df_copy_no_missing_no_id_constant_temp.insert(FactoryID + 1, 'City', failure_df_copy_no_missing_no_id_constant_temp.pop('City'))
failure_df_copy_no_missing_no_id_constant_temp.insert(FactoryID + 1, 'Country', failure_df_copy_no_missing_no_id_constant_temp.pop('Country'))
failure_df_copy_no_missing_no_id_constant_temp.pop("Factory")
```

```python
#Temp variable- remove later
failure_df_copy_no_missing_no_id_constant_temp_split_columns = failure_df_copy_no_missing_no_id_constant_temp
```

```python
failure_df_copy_no_missing_no_id_constant_temp_split_columns.head()
```

### Univariate Analysis of Numerical Exploratory Variables

### Univariate boxplots for Temperature, RPM, Fuel Consumption

```python
plt.figure(figsize=(14,10))

sns.set(style="whitegrid")
plt.subplot(1,3,1)
plt.title("°C")
sns.boxplot(y="Temperature",data=failure_df_copy_no_missing_no_id_constant_temp_split_columns,
            color = "blue",
            showmeans = True,
            meanprops={"marker":"o",
                       "markerfacecolor":"black", 
                       "markeredgecolor":"black",
                      "markersize":"10"} )

plt.ylabel("Average 30 days temperature of the car engine before failure detected")

plt.subplot(1,3,2)
plt.title("RPM")

sns.boxplot(y="RPM",data=failure_df_copy_no_missing_no_id_constant_temp_split_columns,
            color = "red",
            showmeans = True,
            meanprops={"marker":"o",
                       "markerfacecolor":"black", 
                       "markeredgecolor":"black",
                      "markersize":"10"})
plt.ylabel("Average 30 days maximum torque speed at maximum torque before failure detected")
ticks = [x for x in range(0,10,10)]
plt.yticks(ticks)

plt.subplot(1,3,3)
plt.title("Litre/Km")
sns.boxplot(y="Fuel consumption",data=failure_df_copy_no_missing_no_id_constant_temp_split_columns,
            color = "green",
            showmeans = True,
            meanprops={"marker":"o",
                       "markerfacecolor":"black", 
                       "markeredgecolor":"black",
                      "markersize":"10"})
plt.ylabel("Fuel consumption of the car in Litre per kilometre")

plt.show()
```

```python
# Create a new dataframe with only the selected columns
data = failure_df_copy_no_missing_no_id_constant_temp_split_columns[['Temperature', 'RPM', 'Fuel consumption']]

# Calculate the number of outliers in each column using the boxplot method
outliers = data.apply(lambda x: pd.Series(x).quantile(0.75) + 1.5 * (pd.Series(x).quantile(0.75) - pd.Series(x).quantile(0.25)) < pd.Series(x))

# Print the number of outliers in each column
print(outliers.sum())
```

### As seen in the boxplot above, there is significant number of outliers as they are more than 3/2 of upper quartile and 3/2 of the lower quartile

### Correlation matrix for Numerical Exploratory Variables

```python
cat_corr = failure_df_copy_no_missing_no_id_constant_temp_split_columns[['Temperature', 'RPM', 'Fuel consumption']].corr()
half = np.triu(cat_corr)
sns.heatmap(cat_corr, xticklabels=cat_corr.columns.values, yticklabels=cat_corr.columns.values,
            annot=True, vmin=-1, vmax=1, fmt='.1g', cmap='YlGnBu', square=True, mask=half)
sns.set(rc = {'figure.figsize':(12,5)})
```
